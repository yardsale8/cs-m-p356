To fit a model for the values of $x$ in `x_matrix`,

* use the `model.fit(x_matrix, y_vec)` method of the model object.
* `x_matrix` needs to be 2D
* `y_vec` is 1D

#### Predicting the value of $y$ for a given value of $x$


To fit the model for a given value of $x$, we will use `sklearn`'s `predict` method.  To make use of this, we need to set up an x-matrix of values we would like to predict. 



The command is `y_hat = model.predict(x_new)`



We use the $\hat{y}$ to represent the predicted value of $y$ for a given $x$

```python
#Suppose that we want to predict y when x = 5
x = 4.5
#Recall that the x-matrix must be 2D and that we are using a quadratic model (include x and x**2)
y_hat = model.predict([[x, x**2]])
print "The predicted value of $y$ when $x =$", x, "is $\hat{y}$ = ", y_hat[0]
```


###Cross Validation Step 2: Fit a model to each training set, holding out the test fold.

For each fold (called the test set) there is a corresponding train set, which is the rest of the data.

We denote the predicted value of $y^{(i)}$ as $\hat{y}^{(i)}$.  For a regression problem, we will use **mean square prediction error** as the measure of the quality of a model.


$$ MSPE = \frac{\sum_{i=1}^{m}\left( y^{(i)} - \hat{y}^{(i)}\right)^2}{m} $$

```python
#Set the total square prediction error to 0.0
TSPE = 0.0
for train, test in kf:
    #Fit the model to the training data
    model.fit(data[['x','x2']][train],data['y'][train])
    # get the predicted values for the test data.  This returns a vector of y_hat values.
    y_hat = model.predict(data[['x','x2']][test])
    #Compute the differences and sum
    TSPE += np.sum((data['y'][test] - y_hat)**2)
MSPE = TSPE/len(data['y'])
print "The MSPE for the quadratic model was",MSPE,"for this run of cross-validation"
```

#### Comparing different models using cross-validation.

To compare the performance of different models, we

1. Compute the $MSPE$ for each model **using the same folds**
    * Using the same folds reduces variablility
2. Pick the model with the smallest $MSPE$

In a proper study, we would


1. Complete the above steps for many randomly selected folds
2. Average the $MSPE$ for all the trials.

#### Example 1: One set of folds

```python
kf = KFold(len(data.y), n_folds=3, indices=False, shuffle=True)
#Set the total square prediction error to 0.0
linear = linear_model.LinearRegression()
quad = linear_model.LinearRegression()
linear_TSPE = 0.0
quad_TSPE = 0.0
for train, test in kf:
    #Fit the model to the training data
    linear.fit(data[['x']][train],data['y'][train])
    # get the predicted values for the test data
    y_hat = linear.predict(data[['x']][test])
    linear_TSPE += np.sum((data['y'][test] - y_hat)**2)
    #Fit the model to the training data
    quad.fit(data[['x','x2']][train],data['y'][train])
    # get the predicted values for the test data
    y_hat = quad.predict(data[['x','x2']][test])
    quad_TSPE += np.sum((data['y'][test] - y_hat)**2)
linear_MSPE = linear_TSPE/len(data.y)
quad_MSPE = quad_TSPE/len(data.y)
print linear_MSPE, quad_MSPE
```

#### Example 2: Average Over Many Folds

To do this, we add an outer loop and store the MSPE

```python
#Number of times we do k-fold CV
N_trials = 10
#Number of folds
k = 3
linear = linear_model.LinearRegression()
quad = linear_model.LinearRegression()
linear_MSPE = np.zeros(N_trials)
quad_MSPE = linear_MSPE.copy()
for i in range(N_trials):
    kf = KFold(len(data.y), n_folds=k, indices=False, shuffle=True)
    #Set the total square prediction error to 0.0
    linear_TSPE = 0.0
    quad_TSPE = 0.0
    for train, test in kf:
        #Fit the model to the training data
        linear.fit(data[['x']][train],data['y'][train])
        # get the predicted values for the test data
        y_hat = linear.predict(data[['x']][test])
        linear_TSPE += np.sum((data['y'][test] - y_hat)**2)
        #Fit the model to the training data
        quad.fit(data[['x','x2']][train],data['y'][train])
        # get the predicted values for the test data
        y_hat = quad.predict(data[['x','x2']][test])
        quad_TSPE += np.sum((data['y'][test] - y_hat)**2)
    linear_MSPE[i] = linear_TSPE/len(data.y)
    quad_MSPE[i] = quad_TSPE/len(data.y)
print "The average MSPE for the linear model is",linear_MSPE.mean()
print "The average MSPE for the quadratic model is", quad_MSPE.mean()
```

## Cross Validation for Tuning Parameters.

In machine learning, we call a model parameter that isn't part of the hypothesis, but still must be selected, a **tuning parameter**.

Recall that locally weighted linear regression involved a tuning parameter $\tau^2$, which controlled how close a point had to be to get significant weight.



#### Gaussian Weight Function 

$$w_i(x) = \exp\left(-\frac{ (x-x^{(i)})^2 }{ 2\tau^2 } \right)$$

#### Direct Numerical Solution for LLR

\begin{eqnarray}
\vec{\theta} & = & \left(X^TW_xX\right)^{-1}X^TW_x\vec{y}
\end{eqnarray}



It turned out that

* Large $\tau^2$ resulted in a larger weight for far-away points.
* Small $\tau^2$ results in smaller weight for far-away points.

We will select the value of $\tau^2$ using cross-validation.  Again, we will calculate $MSPE$ for each value of $\tau^2$, averaging over many randomly selected folds.



**Note:** Since we are removing a test set, you may run into problems with singular matrices for small $\tau^2$.


```python
#Define a function to compute the weight matrix.
def get_W(x_vec, x_0, tau_sqr):
    ''' Computes W_x where x_vec is the feature numpy vector and x_0 (float) is the value being predicted
        tau_sqr is the band width
    '''
    w_vec = np.exp(-1.0*(x_0 - x_vec)**2/(2*tau_sqr))
    w_mat = np.diag(w_vec)
    return w_mat
```

```python
data2 = pd.read_csv('chwirut2.csv')
#We need the design matrix, so add a column of ones.
data2['ones'] = 1
#Number of times we do k-fold CV
N_trials = 10
#Number of folds
k = 3
# Pick a grid of values for tau^2.  You should start by changing the order of magnitude
tau_list = [0.01, 0.05, 0.1, 1, 5, 10]
MSPE_list = [np.zeros(N_trials) for i in range(len(tau_list))]
for i in range(N_trials):
    kf = KFold(len(data2.y), n_folds=k, indices=False, shuffle=True)
    #Set the total square prediction error to 0.0
    for j, tau in enumerate(tau_list):
        MSPE = 0.0
        for train, test in kf:
            #Make views of the test data.
            x_test = data2['x'][test]
            y_test = list(data2['y'][test].copy())
            #Set up the X and Y matrices using the training sets
            Xmat = np.matrix(data2[['ones','x']][train])
            Ymat = np.matrix(data2[['y']][train])
            for i, x in enumerate(x_test):
                W_i = get_W(data2['x'][train], x , tau)
                theta = (Xmat.T *W_i*Xmat).I*Xmat.T*W_i*Ymat
                y_hat = theta[0,0] + theta[1,0]*x
                error = (y_hat - y_test[i])**2
                MSPE += error/float(len(data2.y))
        MSPE_list[j] += MSPE/N_trials
for i, tau in enumerate(tau_list):
    print "The MSPE for tau^2=",tau,"is",MSPE_list[i].mean()
```

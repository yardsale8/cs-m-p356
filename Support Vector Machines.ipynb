### Maximum Margin Separating Hyperplane

**Support Vector Machines** attempt to maximize the margin, which is pictured below.


The optimization problem, as discussed in [Lecture 6](http://youtu.be/qyyJKd-zXRE) and [Lecture 7](http://youtu.be/s8B4A5ubw6c), is given by

$$ \max_{\gamma, \vec{w}, b} \frac{1}{2}\|\vec{w}\|^2 $$

subject to $y^{(i)}\left(\vec{w}^T\vec{x} + b\right) \ge 1$

The equivalent (dual) problem is

$$ \max_{\vec{\alpha}} \sum_{i=1}^m \alpha_i^2 - \sum_{i=1}^m\sum_{j=1}^m\alpha_i\alpha_jy^{(i)}y^{(j)}\langle \vec{x}^{(i)}\vec{x}^{(j)}\rangle $$

subject to $\alpha \ge 0$ and $\sum_{i=1}^m \alpha_iy^{(i)} = 0$.

Note: This code example was taken from [the `sklearn` website](http://scikit-learn.org/stable/auto_examples/svm/plot_separating_hyperplane.html#example-svm-plot-separating-hyperplane-py).

```python
print(__doc__)

import numpy as np
import pylab as pl
%matplotlib inline
from sklearn import svm

# we create 40 separable points
#np.random.seed(0)
X = np.r_[np.random.randn(20, 2) - [2, 2], np.random.randn(20, 2) + [2, 2]]
Y = [0] * 20 + [1] * 20

# fit the model
clf = svm.SVC(kernel='linear')
clf.fit(X, Y)

# get the separating hyperplane
w = clf.coef_[0]
a = -w[0] / w[1]
xx = np.linspace(-5, 5)
yy = a * xx - (clf.intercept_[0]) / w[1]

# plot the parallels to the separating hyperplane that pass through the
# support vectors
consts = clf.support_vectors_[:,1] - a*clf.support_vectors_[:,0]
b_min = min(consts)
yy_down = a * xx + b_min
b_max = max(consts)
yy_up = a * xx + b_max
sv_ind = [c in (b_min, b_max) for c in consts]

pl.figure(figsize=(10,10))
# plot the line, the points, and the nearest vectors to the plane
pl.plot(xx, yy, 'k-', label="Decision Function")
pl.plot(xx, yy_down, 'k--', label="Margin Boundary")
pl.plot(xx, yy_up, 'k--')

pl.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1],
           s=80, facecolors='none',label="Support Vectors")
pl.scatter(X[:, 0], X[:, 1], c=Y, cmap=pl.cm.Paired)

pl.axis('tight')
pl.title("Linear Decision Boundary with Maximum Margin")
pl.xlabel("Feature 1")
pl.ylabel("Feature 2")
pl.legend()
```

### Soft Margin Classification

As discussed in [Lecture 8](http://youtu.be/bUv9bfMPMb4), we can make adjustments that will allow for some violations of the margin.

(See the images found on [these slides](http://www.robots.ox.ac.uk/~az/lectures/ml/lect2.pdf))


#### Soft Margin Optimization Problem



The optimization problem becomes

$$ \max_{\gamma, \vec{w}, b} \left[\frac{1}{2}\|\vec{w}\|^2 + C\sum_{i=1}^m \xi_i\right] $$

subject to $y^{(i)}\left(\vec{w}^T\vec{x} + b\right) \ge 1 - \xi_i$ and $\xi_i > 0$.


#### Soft Margin Dual Problem
The equivalent (dual) problem is

$$ \max_{\vec{\alpha}} \sum_{i=1}^m \alpha_i^2 - \sum_{i=1}^m\sum_{j=1}^m\alpha_i\alpha_jy^{(i)}y^{(j)}\langle \vec{x}^{(i)}\vec{x}^{(j)}\rangle $$

subject to $0 \le \alpha < c$ and $\sum_{i=1}^m \alpha_iy^{(i)} = 0$.


#### Notes
* The $\xi_i$ are the magnitudes of the margin violations.
* The tuning parameter $C$ adjusts the penalty for a violation.
    * Larger $C$ is a stronger penalty on violation.
    * Larger penalty results in smaller margin.
* Allow for
    * Larger margins
    * Non-seperable data sets

```python
print(__doc__)

import numpy as np
import pylab as pl
%matplotlib inline
from sklearn import svm

# we create 40 separable points
np.random.seed(0)
X = np.r_[np.random.randn(20, 2) - [2, 2], np.random.randn(20, 2) + [2, 2]]
Y = [0] * 20 + [1] * 20

# fit the model
clf = svm.SVC(kernel='linear')
clf.C=0.5
clf.fit(X, Y)

# get the separating hyperplane
w = clf.coef_[0]
a = -w[0] / w[1]
xx = np.linspace(-5, 5)
yy = a * xx - (clf.intercept_[0]) / w[1]

# plot the parallels to the separating hyperplane that pass through the
# support vectors
consts = clf.support_vectors_[:,1] - a*clf.support_vectors_[:,0]
b_min = min(consts)
yy_down = a * xx + b_min
b_max = max(consts)
yy_up = a * xx + b_max
sv_index = np.array([i for i, c in enumerate(consts) if  c == b_min or c == b_max])


pl.figure(figsize=(10,10))
# plot the line, the points, and the nearest vectors to the plane
pl.plot(xx, yy, 'k-', label="Decision Function")
pl.plot(xx, yy_down, 'k--', label="Margin Boundary")
pl.plot(xx, yy_up, 'k--')

pl.scatter(clf.support_vectors_[sv_index, 0], clf.support_vectors_[sv_index, 1],
           s=80, facecolors='none',label="Support Vectors")
pl.scatter(X[:, 0], X[:, 1], c=Y, cmap=pl.cm.Paired)

pl.axis('tight')
pl.title("Linear Decision Boundary with Soft Margin")
pl.xlabel("Feature 1")
pl.ylabel("Feature 2")
pl.legend()
```

### Non-linear Boundaries with Non-linear Kernels



#### The Guassian Kernel

We can get non-linear boundaries by replacing the dot product with a kernel function.  We will focus on the popular ** radius basis function ** or ** Gaussian kernel **

$K(x,y) = \exp\left(-\frac{ \|x-y\|^2 }{ 2\gamma } \right)$

The additional tuning parameter $\gamma$, controls the width of the kernel.

#### Linear Soft Margin Dual Problem
The equivalent (dual) problem is

$$ \max_{\vec{\alpha}} \sum_{i=1}^m \alpha_i^2 - \sum_{i=1}^m\sum_{j=1}^m\alpha_i\alpha_jy^{(i)}y^{(j)}\langle \vec{x}^{(i)}\vec{x}^{(j)}\rangle $$

subject to $0 \le \alpha < c$ and $\sum_{i=1}^m \alpha_iy^{(i)} = 0$.

#### Non-linear Soft Margin Dual Problem with Kernel $K$
The equivalent (dual) problem is

$$ \max_{\vec{\alpha}} \sum_{i=1}^m \alpha_i^2 - \sum_{i=1}^m\sum_{j=1}^m\alpha_i\alpha_jy^{(i)}y^{(j)}K\left( \vec{x}^{(i)},\vec{x}^{(j)}\right)$$

subject to $0 \le \alpha < c$ and $\sum_{i=1}^m \alpha_iy^{(i)} = 0$.


```python
import numpy as np
import pylab as pl
import pandas as pd
from sklearn import svm

data = pd.read_csv('pima_diabetes.csv')
# We need the X matrix to be 2D, but DON'T need to include a column of 1's
X = data[['ogtt','BMI']][data['ogtt'] > 0][data['BMI'] > 0]
# The Y vector must be 1D
Y = data['class'][data['ogtt'] > 0][data['BMI'] > 0]

#SVC stands for support vector classification
clf = svm.SVC(kernel='rbf',C=10,gamma=0.005)

# we create an instance of Neighbours Classifier and fit the data.
clf.fit(X, Y)

# Plot the decision boundary. For that, we will asign a color to each
# point in the mesh [x_min, m_max]x[y_min, y_max].
x_min, x_max = X.min()[0] - .5, X.max()[0] + .5
y_min, y_max = X.min()[1] - .5, X.max()[1] + .5
XX, YY = np.meshgrid(np.linspace(x_min, x_max, 300), np.linspace(y_min, y_max, 300))
Z = clf.predict(np.c_[XX.ravel(), YY.ravel()])


# Put the result into a color plot
Z = Z.reshape(XX.shape)
pl.figure(1, figsize=(10, 10))
pl.pcolormesh(XX, YY, Z, cmap=pl.cm.Paired)


# Plot also the training points
pl.scatter(X['ogtt'], X['BMI'], c=Y, edgecolors='k', cmap=pl.cm.Paired)
pl.xlabel('OGTT')
pl.ylabel('BMI')
pl.xlim(x_min - 0.5, x_max + 0.5)
_ = pl.ylim(y_min - 0.5, y_max + 0.5)
```

## Support Vector Regression

We can also adapt the support vector machine to regression problems.  Here is an example found on the [`sklearn` website](http://scikit-learn.org/stable/auto_examples/svm/plot_svm_regression.html#example-svm-plot-svm-regression-py)

```python
print(__doc__)

###############################################################################
# Generate sample data
import numpy as np

X = np.sort(5 * np.random.rand(40, 1), axis=0)
y = np.sin(X).ravel()

###############################################################################
# Add noise to targets
y[::5] += 3 * (0.5 - np.random.rand(8))

###############################################################################
# Fit regression model
from sklearn.svm import SVR

svr_rbf = SVR(kernel='rbf', C=1e3, gamma=0.1)
svr_lin = SVR(kernel='linear', C=1e3)
svr_poly = SVR(kernel='poly', C=1e3, degree=2)
y_rbf = svr_rbf.fit(X, y).predict(X)
y_lin = svr_lin.fit(X, y).predict(X)
y_poly = svr_poly.fit(X, y).predict(X)

###############################################################################
# look at the results
import pylab as pl
pl.figure(figsize=(10,10))
pl.scatter(X, y, c='k', label='data')
pl.hold('on')
pl.plot(X, y_rbf, c='g', label='RBF model')
pl.plot(X, y_lin, c='r', label='Linear model')
pl.plot(X, y_poly, c='b', label='Polynomial model')
pl.xlabel('data')
pl.ylabel('target')
pl.title('Support Vector Regression')
pl.legend()
pl.show()
```
